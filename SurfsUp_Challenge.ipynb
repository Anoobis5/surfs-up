```python
# Dependencies
import numpy as np
import pandas as pd
import datetime as dt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

#visualizer
from matplotlib import style
style.use('seaborn')
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==6)
#print(june_results)
```

```python
#  3. Convert the June temperatures to a list.
june_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==6).all()
print(june_results)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 

june_df = pd.DataFrame(june_results, columns=['date', 'June Temps'])

# Set the index to date column and ensure correct data query
june_df
```

```python
# Print the Dataframe without the index
#print(june_df.to_string(index=False))
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
print(june_df.describe().to_string(index=True))
# DataFrame format
june_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==12)
#print(dec_results)
```

```python
# 7. Convert the December temperatures to a list.
dec_results = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==12).all()
print(dec_results)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_df = pd.DataFrame(dec_results, columns=['date', 'Dec Temps'])

# Set the index to date column and ensure correct data query
dec_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
print(dec_df.describe().to_string(index=True))
# DataFrame format
dec_df.describe()
```

```python
june_stats = june_df.describe()
dec_stats = dec_df.describe()

summary_df = june_stats.merge(dec_stats, left_index=True, right_index=True)
summary_df
```

### Precipitation

```python
# June Precipitation Measurement

results_prcp_june =[]
results_prcp_june = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()

df_june_prcp = pd.DataFrame(results_prcp_june, columns = ['date', 'June Precipitation'])
df_june_prcp.set_index(df_june_prcp['date'], inplace=True)
```

```python
# June stats
df_june_prcp.describe()
```

```python
results_prcp_jan =[]
results_prcp_jan = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 1).all()

df_jan_prcp = pd.DataFrame(results_prcp_jan, columns = ['date', 'Jan Precipitation'])
df_jan_prcp.set_index(df_jan_prcp['date'], inplace=True)
```

```python
df_jan_prcp.describe()
```

```python
# December Precipitation Measurement
results_prcp_dec =[]
results_prcp_dec = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()

df_dec_prcp = pd.DataFrame(results_prcp_dec, columns = ['date', 'Dec Precipitation'])
df_dec_prcp.set_index(df_dec_prcp['date'], inplace=True)
```

```python
# December Stats
df_dec_prcp.describe()
```

```python
# Merge the two tables to make it more readable and comparable.

june_prcp = df_june_prcp.describe()
dec_prcp = df_dec_prcp.describe()

prcp_summary_df = june_prcp.merge(dec_prcp, left_index=True, right_index=True)
prcp_summary_df
```

### Query Station Data

```python
# June Station Query
june_station = session.query(Measurement.tobs, Measurement.station).filter(extract('month',Measurement.date)== 6).all()
#print(june_station)
```

```python
june_station_df = pd.DataFrame(june_station, columns=['June Temps', 'Station'])
june_station_df
```

```python
june_station_df.tail(10)
```

```python
# December station query

dec_station = session.query(Measurement.tobs, Measurement.station).filter(extract('month',Measurement.date)== 12).all()
#print(dec_station)
```

```python
dec_station_df = pd.DataFrame(dec_station, columns=['Dec Temps', 'Station'])
dec_station_df
```

```python
dec_station_df.tail(10)
```

```python
# Station Data Side-By-Side

station_temps = june_station_df.merge(dec_station_df, on="Station", how="inner").set_index("Station")
station_temps
```
